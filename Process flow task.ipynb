# Getting the List of Rare Orphan Diseases

There is website called https://www.orphadata.com/alignments/ that provides data related to rare disesaes for free. API access is also provided by them but it takes time to get credentials. However, they provide free database in the form of XML, JSON to download for free.

Why this website?

In this data the diseases are mapped in the highearchy so that we are also capable of understanding it's family.

Download Json File at https://www.orphadata.com/alignments/ 

```python
import json 
f = open('en_product1.json',mode='r')
data = json.load(f)
f.close()
```

```python
disorder_name = []
orpha_code = []
disorder_summary = []
for i in range(int(data['JDBOR'][0]['DisorderList'][0]['count'])):
    disorder_name.append(data['JDBOR'][0]['DisorderList'][0]['Disorder'][i]['Name'][0]['label'])
    orpha_code.append(data['JDBOR'][0]['DisorderList'][0]['Disorder'][i]['OrphaCode'])
    try:
        disorder_summary.append(data['JDBOR'][0]['DisorderList'][0]['Disorder'][i]['SummaryInformationList'][0]['SummaryInformation'][0]['TextAuto'][0]['Info'][0]['label'])
    except:
        try:
            disorder_summary.append(data['JDBOR'][0]['DisorderList'][0]['Disorder'][i]['SummaryInformationList'][0]['SummaryInformation'][0]['TextSectionList'][0]['TextSection'][0]['Contents'])
        except:
            disorder_summary.append('No Info')
```

```python
import pandas as pd
df_disease_list = pd.DataFrame()

df_disease_list['Orpha Code'] = orpha_code
df_disease_list['Disorder Name'] = disorder_name
df_disease_list['Summary'] = disorder_summary
```

It is possible that the naming might be slightly different for example 'Aarskog Syndrome' and 'Aarskog-Scott syndrome' are same disease but different sources write differently. Here is the search engine to find the disease in the database based on orphan code or disease name.

```python
#Run this cell to find the disease in the database.

from difflib import get_close_matches
def closeMatches(patterns, word):
    print(get_close_matches(word, patterns))
    
mode = input('Do you want to search by Name or Orphacode? Enter N for Name and O for Orphacode.\n')
if mode=='N':
    name = input('Enter the Name of the Disease.\n')
    name_in_db = get_close_matches(name, df_disease_list['Disorder Name'].tolist())[0]
    print('Orphacoe : ',df_disease_list[df_disease_list['Disorder Name']==name_in_db]['Orpha Code'].values[0])
    print('Name : ',name_in_db)
    print('Summary :', df_disease_list[df_disease_list['Disorder Name']==name_in_db]['Summary'].values[0])
else:
    code = input('Enter the Orphacode of the Disease.\n')
    try:
        print('Orphacoe : ',code)
        print('Name : ',df_disease_list[df_disease_list['Orpha Code']==code]['Disorder Name'].values[0])
        print('Summary :', df_disease_list[df_disease_list['Orpha Code']==code]['Summary'].values[0])
    except:
        print('Code Not found.')
```

# Prevalance Of the Disease

Web Scrapping From Rare Diseases Website

```python
page_names = []
for name in names:
    s=''
    name=name.lower()
    for l in name:
        if l.isalnum():
            s+=l
        else:
            s+='$'
    lst = s.split('$')
    while '' in lst:
        lst.remove('')
    page_names.append('-'.join(lst))
```

```python
import requests
from bs4 import BeautifulSoup
prevalance = []
for name in page_names:
    try:
        url = 'http://worldagnetwork.com/'
        headers = {'User-Agent': 'Chrome/50.0.2661.102'}
        result = requests.get(url, headers=headers)
        # Making a GET request
        r = requests.get('https://rarediseases.org/rare-diseases/'+name+'/#affected-populations',headers=headers)

        # Parsing the HTML
        soup = BeautifulSoup(r.content, 'html.parser')

        s = soup.find('div', class_='wpcf-field-wysiwyg wpcf-field-rd_affected_populations')
        content = s.find_all('p')

        line = ''
        for j in range(len(content)):
            line += ' ' + str(content[j])[3:-4]
        prevalance.append(line)
    except:
        prevalance.append('Error')
```

# Clinical Analysis.gov Data Scrapping

```python
from pytrials.client import ClinicalTrials
df_final = pd.DataFrame()
for search in names[:50]:
    print('Searching for:',search)
    ct = ClinicalTrials()

    # Get 50 full studies related to Coronavirus and COVID in json format.
    ct.get_full_studies(search_expr=search, max_studies=50)

    # Get the NCTId, Condition and Brief title fields from 500 studies related to Coronavirus and Covid, in csv format.
    corona_fields = ct.get_study_fields(
        search_expr=search,
        fields=["NCTId", "Condition","Phase","StudyType", "OverallStatus","LastKnownStatus","IsFDARegulatedDrug",'InterventionType',"LeadSponsorClass","LeadSponsorName","CompletionDate"],
        max_studies=500,
        fmt="csv",
    )

    # Get the count of studies related to Coronavirus and COVID.
    # ClinicalTrials limits API queries to 1000 records
    # Count of studies may be useful to build loops when you want to retrieve more than 1000 records

    ct.get_study_count(search_expr=search)

    # Read the csv data in Pandas
    import pandas as pd

    df=pd.DataFrame.from_records(corona_fields[1:], columns=corona_fields[0])

    def get_years_after_completion(x):
        try:
            year = 2022 - int(str(x[-4:]))
        except:
            year = -1
        return year
    df['ComletedAgoYears']=df['CompletionDate'].apply(get_years_after_completion)

    #Our Constraints
    phases = ['Phase 2', 'Phase 1|Phase 2','Phase 3', 'Phase 2|Phase 3']
    study_types = ['Interventional']
    sponser_type = ['INDUSTRY']
    def isdrug(x):
        if 'Drug' in x.split('|'):
            return True
        else:
            return False
    df['isdrug']=df['InterventionType'].apply(isdrug)
    #Filtering according to our constraints.
    df_out = df[df.Phase.isin(phases) & df.StudyType.isin(study_types) & df.LeadSponsorClass.isin(sponser_type) & (df.ComletedAgoYears<=2)]
    print('Total Number of trails that satisfies the requirement for the selected disease/drug :',df_out.shape[0])
    df_final = pd.concat([df_final,df_out])
```
